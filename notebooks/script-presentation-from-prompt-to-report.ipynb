```python
from openai import OpenAI
from IPython.display import Markdown

client = OpenAI()

MODEL = "gpt-4o-mini"
SYS_MSG = "You are a helpful research assistant"
TEMPERATURE = 0

def get_response(prompt_question):
    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "system", "content": SYS_MSG},
                  {"role": "user", "content": prompt_question}],
        temperature=TEMPERATURE
    )
    
    return response.choices[0].message.content


prompt = "Write me a research report on the benefits of exercise and meditation."
output_report = get_response(prompt)

Markdown(output_report)
```

```python
prompt = "Write me a research report on the benefits of exercise and meditation. Make sure to include a citation to the original reference for each statement made in the report to allow for easy cross checking."
output_report = get_response(prompt)
Markdown(output_report)
```

We can't access actual current information.

```python
prompt = "Find me the latest papers about the benefits of exercise and meditation."

output = get_response(prompt)

Markdown(output)
```

```python
from langchain.agents import create_tool_calling_agent
from langchain_openai import ChatOpenAI
from langchain import hub
from langchain.tools import tool
from langchain.agents import AgentExecutor
from serpapi.google_search import GoogleSearch
import os
import getpass

prompt = hub.pull("hwchase17/openai-tools-agent")

@tool
def web_search(query):
    """
    Perform a web search using the SerpAPI Google Search API.
    """
    serpapi_params = {
    "engine": "google",
    "api_key": os.getenv("SERPAPI_KEY") or getpass("SerpAPI key: ")
    }
    search = GoogleSearch({
        **serpapi_params,
        "q": query,
        "num": 5
    })

    return search.get_dict()["organic_results"]

tools = [web_search]
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

agent = create_tool_calling_agent(llm, tools, prompt)

agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
output = agent_executor.invoke({"input": "Create a research report about the benefits\
    of meditation and exercise. Make sure to include up to date references\
        Use the web_search tool when necessary to guaranteed the validity\
            of all the statements made in the report.",
                             "agent_scratchpad": []})
Markdown(output["output"])
```
